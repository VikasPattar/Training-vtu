```python
import matplotlib.pyplot as plt
 # x axis values

x=[1,2,3]

#corresponding values

y= [2,4,1]

#plotting the points

plt.plot(x,y)

#naming the x axis

plt.xlabel('X - axis')

plt.ylabel('Y - axis')

plt.title('Line Graph')

plt.show()
```

```python
import numpy as np
x= np.arange(0, 10, 0.1)
y= np.sin(x)

plt.figure(figsize = (10, 3))
plt.plot(x,y)
plt.show()
```
