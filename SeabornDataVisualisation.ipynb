```python
sns.set_style('darkgrid')
fig1, axes = plt.subplots(nrow=2,ncols=2, figsize=(14,14))
x=np.random.normal(1,10,1000)
axes[0,0].set_title('Simple KDE Plot')
sns.kdeplot(x, ax=axes[0,0])
axes[0,1].set_title('KDE Plot (Shaded Area under Curve)')
sns.kedplot(x, ax=axes[0,1], shade = True)
axes[1,0].set_title('KDE plot (Differents Color)')
sns.kdeplot(x, ax=axes[1,0], shade=True, color='r', cut = 0)
axes[1,1].set_title('KDE plot (Density on vertical axis)')
sns.kdeplot(x, vertical = True)
plt.show()
```

```python
#Countplot
plt.figure(figsize=(6,9))
sns.countplot(y= 'body-style', data = df, order = df['body-style'].value_counts().index)

plt.show()
```

```python
#boxplot - usually used to dind outliers !
sns.set_style('white')
plt.figure(figsize=(8,6))
sns.boxplot(df['engine-size'])
```

```python
plt.subplots(figsize=(14,24))
plt.subplot(3,1,1)
sns.boxplot(x=df['body-style'], y=df['engine-size'], hue = df['fuel-type'])
plt.subplot(3,1,2)
sns.boxplot(x=df['body-style'], y=df['wheel-base'], hue = df['fuel-type'])
plt.subplot(3,1,3)
sns.boxplot(x=df['body-style'], y=df['width'], hue=df['fuel-type'])
plt.show()
```

```python
#Boxen plot
plt.figure(figsize=(10,9))
sns.boxenplot(x=df['body-style'], y=df['engine-size'], palette = 'Set1')
plt.show()
```

## Bivariate Analysis

```python
# Scatter plot
plt.figure(figsize=(10,10))
sns.scatterplot(x='engine-size', y='wheel-base', data= df, color='#66785F', marker ='o')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='engine-size', y='wheel-base', data=df, color='#fff000', marker ='>', hue ='fuel-type', style='num-of-cylinders')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x='engine-size', y='wheel-base', data=df, hue='fuel-type', size = 'city-mpg', sizes=(20,200))
plt.show()
```

```python
#Line chart
sns.lineplot(x='engine-size', y='wheel-base', data=df)
plt.show()
```

```python
#Regplot - Regression Plot
plt.figure(figsize=(9,9))
sns.regplot(x=df['compression-ratio'], y=df['engine-size'], color='#ff6600', logx=True)
plt.show()
```

```python
sns.set_style('white')
plt.figure(figsize=(16,8))
sns.violinplot(x=df['body-style'], y=df['city-mpg'], palette ='coolwarm', scale='count')
plt.show()
```

```python
# Pair plot
sns.pairplot(df, hue='fuel-type', palette='husl', size=2)
plt.show()
```

```python
sns.pairplot(df, hue='fuel-type', x_vars=['wheel-base','length'], y_vars = ['height','width'], 
             markers=['o','D'], height =3, 
             aspect=1)
plt.show()
```

```python
#Pairgrid
g= sns.PairGrid(df,hue='fuel-type', palette='husl')
g=g.map_offdiag(plt.scatter)
g=g.map_diag(plt.hist, histtype='step', linewidth=2)
g=g.add_legend()
plt.show()
```

```python
#Jointplot
sns.set(style='white', color_codes = True)
g = sns.jointplot(x='wheel-base', y= 'engine-size', data=df, height = 10, color ='#ff6600')
plt.show()
```

```python
g = sns.jointplot(x='wheel-base', y= 'engine-size', data=df, height = 10, kind='hex')
```

```python
g = sns.jointplot(x='wheel-base', y= 'engine-size', data=df, height = 10,
                  color='#ff6600', kind='kde',
                 marginal_kws = {'lw' : 5})
```

```python
#Corr Matrix :
num_values = df.select_dtypes(include=[float, int])
corr = num_values.corr()
corr
```

```python
plt.figure(figsize=(20,10))
ax = sns.heatmap(corr, cmap="YlGnBu", linewidths=0.5, annot=True)
plt.yticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(20,10))
ax = sns.heatmap(corr, cmap="YlGnBu", linewidths=0.5, annot=True, fmt='.1f')
plt.yticks(rotation=0)
plt.show()
```
